```python
import matplotlib.pyplot as plt
```

```python
x=range(1,8)
y=[17,17,18,15,11,11,13]
plt.plot(x,y,color='green',alpha=0.5,linestyle='--',linewidth=3,marker='o',markerfacecolor='blue')
plt.figure(figsize=(20,8),dpi=80)
plt.show()
```
